```python
# Function to extract relevant data from the dictionary
def extract_top_data(overview_str):
    # Parse the string into a dictionary
    overview_dict = ast.literal_eval(overview_str)

    # Extract the "top" part of the dictionary
    top_data = overview_dict.get('top', [])

    # Create a dictionary from the key-value pairs, ignoring 'icon'
    overview_data = {item['key']: item['value'] for item in top_data if 'key' in item and 'value' in item}

    return overview_data

# Extract data from each entry in new_car_overview
overview_data_list = [extract_top_data(car) for car in new_car_overview]

# Convert the list of dictionaries into a DataFrame
all_new_car_overview = pd.DataFrame(overview_data_list)

# Display the resulting DataFrame
print(all_new_car_overview)
```

```python
def extract_feature_data(feature_str):
    # Parse the string into a dictionary
    feature_dict = ast.literal_eval(feature_str)

    # Initialize a dictionary to store all the features
    feature_data = {}

    # Process the 'top' part of the dictionary (if only 'heading' exists)
    top_data = feature_dict.get('top', [])
    feature_data['Features'] = ', '.join(item['value'] for item in top_data)

    # Process the 'data' part of the dictionary (if both 'heading' and 'subHeading' exist)
    for section in feature_dict.get('data', []):
        subheading = section.get('subHeading', section['heading'])  # Use subHeading or heading as the key
        feature_data[subheading] = ', '.join(item['value'] for item in section.get('list', []))

    return feature_data

# Extract data from each entry in new_car_feature
feature_data_list = [extract_feature_data(feature) for feature in new_car_feature]

# Convert the list of dictionaries into a DataFrame
all_new_car_feature = pd.DataFrame(feature_data_list)

# Display the resulting DataFrame
print(all_new_car_feature)
```

```python
# Function to extract relevant data from the dictionary
def extract_spec_data(spec_str):
    # Parse the string into a dictionary
    spec_dict = ast.literal_eval(spec_str)

    # Initialize a dictionary to store all the specifications
    spec_data = {}

    # Process the 'top' part of the dictionary (if only 'heading' exists)
    top_data = spec_dict.get('top', [])
    for item in top_data:
        spec_data[item['key']] = item['value']

    # Process the 'data' part of the dictionary (if both 'heading' and 'subHeading' exist)
    for section in spec_dict.get('data', []):
        subheading = section.get('subHeading', section['heading'])  # Use subHeading or heading as the key
        for item in section.get('list', []):
            spec_data[item['key']] = item['value']

    return spec_data

# Extract data from each entry in new_car_specs
spec_data_list = [extract_spec_data(spec) for spec in new_car_specs]

# Convert the list of dictionaries into a DataFrame
all_new_car_specs = pd.DataFrame(spec_data_list)

# Display the resulting DataFrame
print(all_new_car_specs)
```

Data Handling & Cleaning: Missing Values,Standardising  Formats,Encoding Categorical Variables,Removing Outliers & Exploratory Data Analysis (EDA)

```python
import pandas as pd
import numpy as np
from sklearn.preprocessing import MinMaxScaler
from sklearn.ensemble import RandomForestRegressor
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import LabelEncoder
from scipy import stats
from io import StringIO
import matplotlib.pyplot as plt
import seaborn as sns
import re
import os
import joblib
import time

folder_path = '/content/'
file_name = 'All_Cars_detail.xlsx'
file_path = os.path.join(folder_path, file_name)
# Load the dataset
carfinal = pd.read_excel('/content/All_Cars_detail.xlsx')

print( carfinal.shape)
carfinal.head()
```

```python
# 1) In price column Replace ₹ symbol & convert crore , lakh value into the standard Rs value
carfinal['price'] = carfinal['price'].replace({'₹': '', ',':''}, regex=True)
def convert_to_thousand(price):
    price = str(price).lower().strip()  # Convert to string, lowercase, and strip spaces
    if 'crore' in price:
        return float(price.replace('crore', '').strip()) * 10000000
    elif 'lakh' in price:
        return float(price.replace('lakh', '').strip()) * 100000
    else:
        try:
            return float(price)  # Assume it's already in Thousand if no keyword is present
        except ValueError:
            return None  # Handle unexpected cases by returning None

# Apply the conversion function to the Price column
carfinal['price'] = carfinal['price'].apply(convert_to_thousand)

# Optionally, convert the Price column to float
carfinal['price'] = carfinal['price'].astype(float)


# 2) Remove the duplicate columns 'owner','Ownership' , 'Kms Driven','Seats.1','Seats','Engine Displacement' column from the carfinal DataFrame
carfinal.drop(columns=['owner', 'Ownership', 'Kms Driven', 'Seats.1','Transmission', 'Seats','Engine Displacement','Fuel Type'], inplace=True, errors='ignore')


# 3) Identify columns that have the same value across all rows (constant columns)
constant_columns = [col for col in carfinal.columns if carfinal[col].nunique() == 1]
# Drop the constant columns from the DataFrame
carfinal = carfinal.drop(columns=constant_columns)



# 4) Calculate the number of missing values in each column
missing_values_count = carfinal.isnull().sum()
# Calculate the percentage of missing values in each column
missing_values_percentage = (carfinal.isnull().sum() / len(carfinal)) * 100
# Create a DataFrame to store the missing value information
missing_values_df = pd.DataFrame({
    'Missing Values Count': missing_values_count,
    'Missing Values Percentage': missing_values_percentage})
# Remove columns with more than 10 missing values, keeping 'Mileage' regardless of its missing values
columns_to_keep = [col for col in carfinal.columns if missing_values_count[col] <= 10 or col == 'Mileage']
# Update the DataFrame to keep only the selected columns
carfinal = carfinal[columns_to_keep]


# 5) Remove non-numeric characters from 'Mileage' column and convert to float
if carfinal['Mileage'].dtype != 'float64':  # Check if 'Mileage' is already float
    carfinal.loc[:, 'Mileage'] = pd.to_numeric(carfinal['Mileage'].str.extract('(\d+\.?\d*)')[0], errors='coerce')



# 6) Remove commas from 'km' column and convert to int
if not pd.api.types.is_integer_dtype(carfinal['km']):  # Check if 'km' is already int
    carfinal.loc[:, 'km'] = pd.to_numeric(carfinal['km'].str.replace(',', ''), errors='coerce').astype('Int64')



# 7) List of columns to be converted to string and then category
columns_to_convert = ['ft', 'bt', 'transmission', 'oem', 'model', 'City']
# Convert columns to string, then category
for col in columns_to_convert:
    carfinal[col] = carfinal[col].astype(str).astype('category')


# 8) List of columns to be converted to numeric format
columns_to_convert_to_numeric = ['km', 'ownerNo', 'Mileage', 'Seating Capacity', 'modelYear']
# Convert columns to numeric
for col in columns_to_convert_to_numeric:
    carfinal[col] = pd.to_numeric(carfinal[col], errors='coerce')


# 9) Remove irrelavant column from the carfinal DataFrame
carfinal = carfinal.copy()  # Make a copy if you suspect it's a view
carfinal.drop(columns=['Displacement', 'car_links', 'centralVariantId', 'variantName','Insurance Validity','Color','Engine'], inplace=True, errors='ignore')


# 10) Fill missing values in 'Mileage' column with its median
carfinal['Mileage'].fillna(carfinal['Mileage'].median(), inplace=True)
 # Fill missing values in 'Seating Capacity' column with its median
carfinal['Seating Capacity'].fillna(carfinal['Seating Capacity'].median(), inplace=True)


# Save the cleaned data
output_file_path = os.path.join(folder_path, 'carfinal.csv')
carfinal.to_csv(output_file_path, index=False)
print(f"DataFrame saved to: {output_file_path}")

# Display the first few rows of the modified DataFrame
print(carfinal.head(1))
print( carfinal.shape)

# Check for missing values in all columns of the carfinal DataFrame
missing_values = carfinal.isnull().sum()

# Print the missing values for each column
print(missing_values)
```

```python
#Normalizing numerical features using Min-Max Scaling, excluding 'price'
scalers = {}
def min_max_scaling(carfinal, columns):
    global scalers
    for col in columns:
        if col in carfinal.columns:
            scaler = MinMaxScaler()
            carfinal[col] = scaler.fit_transform(carfinal[[col]])
            scalers[col] = scaler
    return carfinal

carfinal = min_max_scaling(carfinal,  ['km', 'ownerNo', 'Mileage', 'Seating Capacity', 'modelYear'])

# Label Encoding for categorical variables
label_encoders = {}
def label_encode(carfinal, columns):
    global label_encoders
    for col in columns:
        if col in carfinal.columns:
            le = LabelEncoder()
            carfinal[col] = le.fit_transform(carfinal[col].astype(str))
            label_encoders[col] = le
    return carfinal
categorical_columns = ['ft', 'bt', 'transmission', 'oem', 'model', 'City']
carfinal = label_encode(carfinal, categorical_columns)

# Identify and Removing outliers using IQR for 'price'
Q1 = carfinal['price'].quantile(0.25)
Q3 = carfinal['price'].quantile(0.75)
IQR = Q3 - Q1
carfinal = carfinal[(carfinal['price'] >= (Q1 - 1.5 * IQR)) & (carfinal['price'] <= (Q3 + 1.5 * IQR))]

# Save preprocessing steps
joblib.dump(label_encoders, '/content/label_encoders.pkl')
joblib.dump(scalers, '/content/scalers.pkl')

print(carfinal.columns.tolist())
print( carfinal.shape)
```
